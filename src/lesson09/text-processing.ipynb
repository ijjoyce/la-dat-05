```python
from sklearn.datasets import fetch_20newsgroups 
categories = [
    'alt.atheism', 
    'soc.religion.christian', 
    'comp.graphics', 
    'sci.med'
]
twenty_train = fetch_20newsgroups(subset='train', categories=categories, 
                                  shuffle=True, random_state=42)
```

```python
print twenty_train.data[0]
```

```python
twenty_train.target[:10]
```

```python
twenty_train.target_names
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
vect = CountVectorizer(stop_words='english', min_df=3, lowercase=True, ngram_range=(1,2))
X_train_counts = vect.fit_transform(twenty_train.data)
```

```python
print X_train_counts[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, TfidfTransformer
```

```python
transformer = TfidfTransformer()
X_train_tfidf = transformer.fit_transform(X_train_counts)
```

```python
print X_train_tfidf[0]
```

```python
reversed_vocab = {v: k for (k, v) in vect.vocabulary_.items()}
```

```python
X_train_tfidf
```

```python
from sklearn.cross_validation import cross_val_score 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.linear_model import LogisticRegression 
from sklearn.naive_bayes import MultinomialNB
```

```python
X_train_tfidf_dense = X_train_tfidf.toarray()
tree = DecisionTreeClassifier()
print cross_val_score(tree, X_train_tfidf_dense, twenty_train.target, cv=5)
```

```python
forest = RandomForestClassifier(n_estimators=10)
cross_val_score(forest, X_train_tfidf_dense, twenty_train.target, cv=3)
```

```python
logreg = LogisticRegression()
cross_val_score(logreg, X_train_tfidf, twenty_train.target)
```

```python
nb = MultinomialNB()
cross_val_score(nb, X_train_tfidf, twenty_train.target)
```
